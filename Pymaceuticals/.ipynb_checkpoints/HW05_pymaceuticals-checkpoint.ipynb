```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"
```

```python
# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
clinic_df = pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset
combined_df = clinic_df.merge(mouse_df, how='left')
# Display the data table for preview
combined_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
tum_mean = combined_df[['Drug','Timepoint','Tumor Volume (mm3)']].groupby(['Drug','Timepoint']).mean()
tum_mean_df = tum_mean.reset_index()
# Preview DataFrame
tum_mean_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
tum_sem = combined_df[['Drug','Timepoint','Tumor Volume (mm3)']].groupby(['Drug','Timepoint']).sem()
tum_sem_df = tum_sem.reset_index()
# Preview DataFrame
tum_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tum_mean_pivot = tum_mean_df.pivot(index='Timepoint', columns='Drug', values="Tumor Volume (mm3)")
# Preview that Reformatting worked
tum_mean_pivot
```

```python
# Create sem table for errorbar graph
tum_sem_pivot = tum_sem_df.pivot(index='Timepoint', columns='Drug', values="Tumor Volume (mm3)")

# Generate the Plot (with Error Bars)
drug_attribute = [['Capomulin','o','r'],['Infubinol','^','b'],['Ketapril','s','g'],['Placebo','d','k']]
for drug,fmt,color in drug_attribute:
    drug_error_bar = plt.errorbar(tum_mean_pivot.index,tum_mean_pivot[drug],tum_sem_pivot[drug],
                                  fmt=fmt,color=color,label=drug,ls='--',linewidth=0.8,alpha=0.8)
# Plot labels and Attributes
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend()
plt.grid()
# Save the Figure
plt.savefig("Images/img1_treatment.png")
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint
meta_mean_df = combined_df[['Drug','Timepoint','Metastatic Sites']].groupby(['Drug','Timepoint']).mean()
meta_mean_df.head() ## I did not reset index just to match given solution
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint
meta_sem_df = combined_df[['Drug','Timepoint','Metastatic Sites']].groupby(['Drug','Timepoint']).sem()
meta_sem_df.head() ## I did not reset index just to match given solution
```

```python
# Format to DataFrame
meta_mean_df = meta_mean_df.reset_index()
meta_sem_df = meta_sem_df.reset_index()
# Minor Data Munging to Re-Format the Data Frames
meta_mean_pivot = meta_mean_df.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')
meta_mean_pivot
```

```python
# Create sem table for errorbar graph
meta_sem_pivot = meta_sem_df.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')

# Generate the Plot (with Error Bars)
for drug,fmt,color in drug_attribute:
    meta_error_bar = plt.errorbar(meta_mean_pivot.index,meta_mean_pivot[drug],meta_sem_pivot[drug],
                                  fmt=fmt,color=color,label=drug,ls='--',linewidth=0.8,alpha=0.8)
# Plot labels and Attributes
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.legend()
plt.grid()
# Save the Figure
plt.savefig("Images/img2_spread.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
mouse_survival_df = combined_df.groupby(['Drug','Timepoint'])['Mouse ID'].nunique()
mouse_survival_df = mouse_survival_df.reset_index().rename(columns={'Mouse ID':'Mouse Count'})
# Preview DataFrame
mouse_survival_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_survival_pivot = mouse_survival_df.pivot(index='Timepoint',columns='Drug',values='Mouse Count')
# Preview the Data Frame
mouse_survival_pivot
```

```python
# Generate the Plot (Accounting for percentages)
for drug,fmt,color in drug_attribute:
    meta_error_bar = plt.plot(mouse_survival_pivot.index,(mouse_survival_pivot[drug]/25)*100,
                              marker=fmt,color=color,label=drug,ls='--',linewidth=0.8,alpha=0.8)
# Plot labels and Attributes
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend()
plt.grid()
# Save the Figure
plt.savefig("Images/img3_survival.png")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
per_change =  ((tum_mean_pivot.iloc[-1]/tum_mean_pivot.iloc[0])-1)*100
# Display the data to confirm
print(per_change)
type(per_change)
```

```python
# Store all Relevant Percent Changes into a Tuple
drug_per_change = []
drug_list = []

for drug,m,c in drug_attribute:
    drug_per_change.append(per_change[drug].astype(np.int64))
    drug_list.append(drug)

tup_drug_pc = tuple(drug_per_change)

# Splice the data between passing and failing drugs

bar_graph = plt.bar(drug_list,tup_drug_pc,width=-1,align='edge',
                    color=['g' if tup_drug_pc[i]<0 else 'r' for i in np.arange(len(tup_drug_pc))])

# Orient widths. Add labels, tick marks, etc. 

plt.title('Tumor Change Over 45 Day Treatment')
plt.ylabel('% Tumor Volume Change')
plt.ylim(-30,70)
plt.xlim(-1,3)
plt.grid()
plt.xticks(ha='right')
plt.tick_params(axis='x',width=0)

# Use functions to label the percentages of changes
def barlabel(num):
    for v in num:
        height = v.get_height()
        if height < 0:
            y_place = -7
        else:
            y_place = 5
            
        plt.text(v.get_x()+v.get_width()/2,y_place,'%d'%int(height)+'%',
                color='w',ha='center')

# Call functions to implement the function calls
barlabel(bar_graph)

# Save the Figure
plt.savefig("Images/img4_change.png")
# Show the Figure
plt.show()

```

## Pymaceutical Observations
#### Written Analysis

1. All plots show that **'Capomulin'** was the only drug with positive results in all aspects. Reduced tumor volume, suppressed spread, and outstanding high survival rate compare to other drugs.

2. **'Infubinol'** did show positive outcome for metastatic spread during treatment, but this treatment led to lowest survival rate over time. 

3. **'Ketapril'** and **'Placebo'** showed no significance in general among four various treatments. Over 45 Day result **'Ketapril'**'s record was no better than **'Placebo'** at all.
